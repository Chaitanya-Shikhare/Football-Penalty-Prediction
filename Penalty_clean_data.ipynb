```python
import pandas as pd
```

```python
df = pd.read_csv("categorical_data.csv" ,encoding='latin-1')
```

```python
df.head()
```

```python
df.describe
```

```python
df.isnull().sum()
```

```python
df["competition_name"] = df["competition_name"].fillna("unknown")
df["competition_stage"] = df["competition_stage"].fillna("unknown")
df["stadium"] = df["stadium"].fillna("unknown")
df["direction"] = df["direction"].fillna("unknown")
df.isnull().sum()
```

```python
df
```

```python
df_new =df.copy()
```

```python
df_new
```

```python
df_new.columns
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
df_new.player_team_name = le.fit_transform(df_new.player_team_name)
df_new.player_name = le.fit_transform(df_new.player_name)
df_new.player_position = le.fit_transform(df_new.player_position)
df_new.foot = le.fit_transform(df_new.foot)
df_new.technique = le.fit_transform(df_new.technique)
df_new.keeper_name = le.fit_transform(df_new.keeper_name)
df_new.keeper_team_name = le.fit_transform(df_new.keeper_team_name)
df_new.competition_name = le.fit_transform(df_new.competition_name)
df_new.competition_stage = le.fit_transform(df_new.competition_stage)
df_new.stadium = le.fit_transform(df_new.stadium)
df_new.direction = le.fit_transform(df_new.direction)
df_new
```
